What is Python ?

Python is a high-level general-purpose programming language known for its simplicity and readability. It was created by Guido van Rossum and first released in 1991.It is an open source programming language. Python emphasizes code readability and allows developers to express concepts in fewer lines of code than other languages like C++ or Java.

Variables : These are used to temporarily store data in Computer's memory. For example, price = 20 . Python Interpreter will allocate some memory and store 20 in the memory and it will attach the price label to that memory location.

Rules for variable names :
1. Variable names must start with a letter or underscore character. It cannot start with a number
2. Always use lower case letters for variable names. It cannot have Capital letters
3. A variable name can contain only alphanumeric characters and underscores
4. variable names are case sensitive
5. Reserved keywords cannot be used as variable names. (example : True, False, print etc)
6. Variable names should be easy to code , easy to understand and easy to debug

Arthimetic Operators

| Operator | Description            | Example  | Result |
| -------- | ---------------------- | -------- | ------ |
| `+`      | Addition               | `5 + 3`  | `8`    |
| `-`      | Subtraction            | `5 - 3`  | `2`    |
| `*`      | Multiplication         | `5 * 3`  | `15`   |
| `/`      | Division (float)       | `5 / 2`  | `2.5`  |
| `//`     | Floor Division         | `5 // 2` | `2`    |
| `%`      | Modulus (remainder)    | `5 % 2`  | `1`    |
| `**`     | Exponentiation (power) | `2 ** 3` | `8`    |




```python
x = 10
x = x + 3
# augmented assignment operator
x += 3
x -= 3

a = 10
b = 3

print(a + b)   # 13
print(a - b)   # 7
print(a * b)   # 30
print(a / b)   # 3.333...
print(a // b)  # 3
print(a % b)   # 1
print(a ** b)


#/ always returns a float, even if the result is a whole number.

#// (floor division) removes the decimal part, rounding down.

#% is often used to check if a number is even or odd.
```

Operator Precedence

| Precedence Level | Operators                                                                 | Description                       |            |
| ---------------- | ------------------------------------------------------------------------- | --------------------------------- | ---------- |
| 1 (highest)      | `()`                                                                      | **Parentheses** – evaluated first |            |
| 2                | `**`                                                                      | Exponentiation                    |            |
| 3                | `+x`, `-x`, `~x`                                                          | Unary plus, minus, bitwise NOT    |            |
| 4                | `*`, `/`, `//`, `%`                                                       | Multiplication, division, etc.    |            |
| 5                | `+`, `-`                                                                  | Addition, subtraction             |            |
| 6                | `<<`, `>>`                                                                | Bitwise shift operators           |            |
| 7                | `&`                                                                       | Bitwise AND                       |            |
| 8                | `^`                                                                       | Bitwise XOR                       |            |
| 9                | \`                                                                        | \`                                | Bitwise OR |
| 10               | Comparison operators (`==`, `!=`, `>`, `<`, `>=`, `<=`, `is`, `in`, etc.) | Comparisons                       |            |
| 11               | `not`                                                                     | Logical NOT                       |            |
| 12               | `and`                                                                     | Logical AND                       |            |
| 13               | `or`                                                                      | Logical OR                        |            |
| 14 (lowest)      | `=`, `+=`, `-=`, etc.                                                     | Assignment operators              |            |
